```python
import requests
import json
from PIL import Image
import urllib.request
import matplotlib.pyplot as plt
```

```python
url = "https://modelslab.com/api/v6/realtime/text2img"
api_key = "XXXXXXXXXXX"
```

```python
# Headers de la petici칩n
headers = {
  'Content-Type': 'application/json'
}
```

```python
# Body de la petici칩n
payload = json.dumps({
    "key" : api_key,
    "prompt": "a woman with hippie style, looking at the camera",
    "negative_prompt": "bad quality",
    "width": "512",
    "height": "512",
    "safety_checker": False,
    "seed": None,
    "samples":1,
    "base64":False,
    "webhook": None,
    "track_id": None
})
```

```python
# Ejecutar petici칩n
response = requests.post(url, data=payload, headers=headers)
data = response.json()
```

```python
# Mostrar im치genes generadas
for image in data['output']:
    img = Image.open(requests.get(image, stream=True).raw)
    plt.imshow(img)
    plt.axis('off')
    plt.show()
```
